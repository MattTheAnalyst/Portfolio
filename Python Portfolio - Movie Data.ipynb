```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Let's read in the data

df = pd.read_csv('/Users/matthewrowe/Desktop/Python Exercises /movies.csv')
```

```python
#Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
pd.set_option('display.float_format', lambda x: '%.f' % x)
pd.options.display.float_format = '{:,}'.format

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Now let's make a scatter plot of Budget vs Gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')
plt.show()
```

```python
df.head()
```

```python
#We'll now add a regression line

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "orange"}, line_kws={"color":"blue"})
```

```python
#Let's show the numeric correlation
#Choosing from pearson, spearman and kendall

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Heatmap')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#What about non-numeric features?
#We can convert to a numeral to represent companies

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```
